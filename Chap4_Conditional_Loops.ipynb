- The not operator on a whole expression can be spread to apply the not one each individual codition.

- not (codition1  and condition2) --> not condition1 or not codition2
- not (codition1 or condition2) --> not condition1 and not condition2

# loops
- We geenrally need to follow a DRY rule when programming.
- DRY - Do not Repeat Yourself.
- This tell us to avoid repeating the code that we write.
- Till now we have been repeating code when we had to do multiple operation which are similar.
- WE can use loops to solve this to a certain extent.
- loop is a programming concept that helps us to do an operation multiple times without copying the code again
- There are 2 types of loops
    1. while loop - infinite loop - work with numbers, strings, lists, dictionaries converted to list
    2. for loop - finite loop - works with numbers using range(), list, strings, dictionaries

# While loops
- while loops are infinite loops.
- There is a possibility for it to become infinite
- Or we can deliberately make it infinite
- There 2 forms of a while loop
    - while loop with a codition - can also become infinite 
    - infinite loop

### while loop with a codition
- WE need 3 compnents to create a while loop with a condition
1. variable with initial value
2. Condition that controls the loop and decides when the loop has to keep running or when to stop
3. Increment or decrement that allows us to move the variable towards the end.

- Syntax:
```python
variableName = value
while codition:
    code1
    code2
    ...
    increment/ decrement
```
- Note: the condition can have multiple coditions aswell and the variables could be multiple variables. There could be multiple increments and recrements
- While loops has a block of code that belongs to it. So the indentation is important.

# Aphine Cipher
- 2 keys are needed for this method.
- formula = (ax + b) mod 26
- a is the first key and b is the second key.
- x is location of the character from a to z that we want to encypt.
- This algorith is a mix of additive and multiplicative cipher.
- First we convert the character to a number based on the alphabets where we start a with 0 and we ends with z having 25.
- we multiply it with the first key.
- Then we add it with the second key 
- Then we divide it with 26 to get the new character's number.
- Then we convert the number into character.
- WE combine the characters to get the encrypted string.
- if we want we can include the spaces and digits and symbols aswell but the mod value  has to be changed.
 

````python
# alpha = 'abcdefghijklmnopqrstuvwxyz'
import string 
# sting module contains uppercase, lowercase character, digits and symbols aswell that we can import and make use of.
alpha = string.ascii_lowercase
key1  = int(input('Enter the key1: '))
key2  = int(input('Enter key2: '))
sentence = input('Enter a sentence: ')
encryptedText = ''
for char in sentence:
    if char in alpha:
        # finding location in the alphabets
        index = alpha.index(char)
        # using the formula
        index *= key1
        index += key2
        index %= len(alpha)
        # converting back to character.
        encryptedText += alpha[index]
    else:
        # if not a character, we simply add the characte to the encrupted text.
        encryptedText += char
print(encryptedText)
````

# Vigenere Cipher
- consider a word as a key.
- make the key same length as the plain text by repeating it.
- for each character in plaintext get number for the plain character and the number for the character in the key.
- add them together
- mod 26
- convert back to character


````python
alpha = string.ascii_lowercase
sentence = input('Enter a sentence.Dont give spaces as the key does not have spaces: ')
# print(len(sentence))
encryptedText = ''
word = input('Enter a word for the key: ')
if len(word)> len(sentence):
    key = word[ : len(sentence)]
else:
    repeat = len(sentence) / len(word)
#     print(repeat)
    key = word * int(repeat)
    remaining = len(sentence) - len(key)
    key += word[ : remaining]
#     print(key)
#     print(len(key))

for char1,char2 in zip(sentence,key):
    num1 = alpha.index(char1)
    num2 = alpha.index(char2)
    newNum = (num1 + num2) %len(alpha)
    encryptedText += alpha[newNum]
print(encryptedText)
````

Returns true if for every '*' (star) in the string, if there are chars both immediately before and after the star, they are the same.

sameStarChar("xy*yzz") → true
sameStarChar("xy*zzz") → false
sameStarChar("xa*az") → true

````python
word = input('Enter a string: ')
same = True
for index in range(len(word)):
    char = word[index]
    if char  == '*':
        if index  != 0 and index != len(word)-1:
            if word[index - 1 ] != word[index + 1]:
                same = False
                break
        else:
            same = False
            break
print(same)
````

Write a program that takes a list of words as input and prints the length of each word, except for words that have a length of 3. Use the continue statement to skip words with a length of 3.

````python
words = {}
numberOfWords = 5
for _ in range(numberOfWords):
    word = input('Enter a word: ')
    if len(word) <= 3:
        continue
    words[word] = len(word)
print(words)
````

# Nested loops
- Loop inside another loop is called a nested loop
- Nested loops are more powerful and more complex.
- They are used when working with multiple lists asynchronously or when we work with a 2d grid system (nested list) for many games like tic tac toe, battle ship, chess and so on.
- if each value in the internal list or dictionary needs to perform an operation, then we use a nested loop.
- The important thing about the nested loops is that the inner loop has to finish before it jumps to the next round of the outer loop.
- When inner loop finishes running the variables for the inner loop gets destroyed.

````python
for i in range(1 , 4): # [1,2,3]  
    for j in range(1, 4): #[1,2,3] 
        print(f'i= {i}, j= {j}')
    print()
print('Finished the whole loop')
````
