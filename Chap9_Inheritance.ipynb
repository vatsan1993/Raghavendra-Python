# Relationships:
- If a class is dependent on another class, there is a relationship between them.
1. has-a-relationship
2. is-a-relationship

# Has a Relationship
- if a class has another class's object as its field or member, then this relationship is called has a relationship
- Example:
    - A Car has an Engine.  (Aggregation)
        - if car is sold out, engines might not be soldout
        - weak relationship
    - A Person will have Address (Composition) 
        - if the person delete his profile, the address has to be deleted aswell.
        - strong relationship

```python
# footballplayer-height
# Note: it is better to use propeties and setters instead of getter and setter
class Height:
    def __init__(self, foot, inches):
        self._foot = foot
        self._inches = inches
    def getFoot(self):
        return self._foot
    def setFoot(self, foot):
        self._foot = foot
    def getInches(self):
        return self._inches
    def setInches(self, inches):
        self._inches = inches
    def __str__(self):
        return f"{self._foot}'{self._inches}''"
    def __repr__(self):
        return f"{self._foot}'{self._inches}''"

class FootballPlayer:
    def __init__(self, name, jersey_num, age, position, height, weight, num_games, num_wins , num_losses, num_score):
        self._name = name
        self._jersey_num = jersey_num
        self._age = age
        self._position = position
        self._height = height
        self._weight = weight
        self._num_games = num_games
        self._num_wins = num_wins
        self._num_losses = num_losses
        self._num_score = num_score
    def calc_rating(self):
        rating = 3.0
        if self._age < 35 and self._num_score > 200:
            rating = 4.2
        elif self._age < 35 and self._num_score > 80:
            rating = 3.5
        elif self._age > 35:
            if self._num_wins > 10 * self._num_losses:
                rating = 4.7
            elif self._num_wins > 5 * self._num_losses:
                rating = 4.1
            elif self._num_wins > 2 * self._num_losses:    
                rating = 3.2
        return rating
    def __str__(self):
        return f'Name= {self._name}, Age= {self._age}, Jersey= {self._jersey_num}, Height= {self._height}, Weight= {self._weight} Rating= {self.calc_rating()}'
    

h = Height(6,2)
player1 = FootballPlayer("Micheal Jordan", 10, 56, "Front-Right", h, 150, 120, 117 , 1, 520)
print(player1)
player2 = FootballPlayer("George", 21, 25, "Back-left", h, 95, 25, 15 , 3, 120)
print(player2)
```

```python
class Meeting:
    def __init__(self, meeting_id, name, description, start_time, duration, total_capacity = 100, price = 20):
        self._meeting_id = meeting_id
        self._name = name
        self._description = description
        self._start_time = start_time
        self._duration = duration
        self._end_time = start_time + timedelta(minutes=duration)
        self._total_capacity = total_capacity
        self._available_tickets = total_capacity
        self._price = price
        self._students = []
        
    def getId(self):
        return self._meeting_id
    def get_name(self):
        return self._name
    def get_description(self):
        return self._description
    def get_price(self):
        return self._price
    def __str__(self):
        return f'''Meeting_id: {self._meeting_id} 
        Name:{self._name}
        Description: {self._description}
        Start time: {self._start_time}
        End time: {self._end_time}
        Duration: {self._duration}
        Available Tickets: {self._available_tickets}
        Price: ${self._price}'''
    def __repr__(self):
        return f'''Meeting_id: {self._meeting_id} 
        Name:{self._name}
        Description: {self._description}
        Start time: {self._start_time}
        End time: {self._end_time}
        Duration: {self._duration}
        Available Tickets: {self._available_tickets}
        Price: ${self._price}'''
    def add_student(self, student):
        self._students.append(student)
        self._available_tickets -= 1
    
    def remove_student_from_meeting(self, student_id):
        for i in range(len(self._students)):
            student = self._students[i]
            if student.getId() == student_id:
                self._students.pop()
                self._available_tickets += 1
                return student
    
    def search_student_in_meeting_by_id(self, student_id):
        for student in self._students:
            if student.getId() == student_id:
                return student
        return None
    
    def search_student_in_meeting_by_phone(self, phone):
        for student in self._students:
            if student.getPhone() == phone:
                return student
        return None
    
    def search_student_in_meeting_by_Name(self, name):
        students = []
        for student in self._students:
            if name  in student.getName():
                students.append(student)
        return students
    def matching_keyword(self, keword):
        if keyword in self._name or keyword in self._description:
            return True
    def get_stundents(self):
        return self._students
```
