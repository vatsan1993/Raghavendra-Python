# Demorgan's Law
- The not operator on a whole expression can be spread to apply the not one each individual codition.
- When we do this the relational operator also change into its opposite operator.
- not (codition1  and condition2) --> not condition1 or not codition2
- not (codition1 or condition2) --> not condition1 and not condition2


````python
sentence = 'Hi, my name is max.'
# print(sentence.startswith('how'))
# print(sentence.endswith('?'))
isNotQuestion = not (sentence.startswith('how') or sentence.endswith('?'))

print(isNotQuestion)
````

# if statement

```python

    ...
```

# if else statement
- Using the if statement we can only run the code when the condition is True.
- But with  if else, we can run code even when the condition if false, but the code will be a different set of operations
- Syntax:
```python
if condition:
    code1
    code2
    ...
else:
    code3
    code4
    ...
```
- if ther condition in the if statement is False it automatically jumps to the else statement.
- else statment will not have its own condition

````python
number = int(input("Enter a number: "))
if number >= 0:
    print("number is positive")
else:
    print('number is negative')
print("if statement finished successfully!")
````

# elif
- If we have more than 2 conditions to check, then we use the elif statements.
- elif statements can have any number of conditions.
- it ends with else block
- else block is optional
- its always better to have the else block
- if a condition is True, python runs that code and exits. It does not checks other conditions.
- if a condition is False, then python checks the next condition.
- Syntax:
```pyhton
if condition1:
    code1
    code2
    ...
elif condition2:
    code3
    code4
    ...
elif condition3:
    code5
    code6
    ...
...
else:
    code n-1
    code n
```

# loops
- We geenrally need to follow a DRY rule when programming.
- DRY - Do not Repeat Yourself.
- This tell us to avoid repeating the code that we write.
- Till now we have been repeating code when we had to do multiple operation which are similar.
- WE can use loops to solve this to a certain extent.
- loop is a programming concept that helps us to do an operation multiple times without copying the code again
- There are 2 types of loops
    1. while loop - infinite loop - work with numbers, strings, lists, dictionaries converted to list
    2. for loop - finite loop - works with numbers using range(), list, strings, dictionaries

# While loops
- while loops are infinite loops.
- There is a possibility for it to become infinite
- Or we can deliberately make it infinite
- There 2 forms of a while loop
    - while loop with a codition - can also become infinite 
    - infinite loop

### while loop with a codition
- WE need 3 compnents to create a while loop with a condition
1. variable with initial value
2. Condition that controls the loop and decides when the loop has to keep running or when to stop
3. Increment or decrement that allows us to move the variable towards the end.

- Syntax:
```python
variableName = value
while codition:
    code1
    code2
    ...
    increment/ decrement
```
- Note: the condition can have multiple coditions aswell and the variables could be multiple variables. There could be multiple increments and recrements
- While loops has a block of code that belongs to it. So the indentation is important.


````python
print('hi')
print('hi')
print('hi')
print('hi')
print('hi')
print('hi')
print('hi')
print('hi')
print('hi')
print('hi')
````

````python
count = 1
end = 100
message = 'hi'
while count <= end:
    print(message, end = " ")
    count +=1 
````

### Inifinite Loop
- The infinite loop runs for ever.
- This needs only 1 component.
- That component is True.
- Syntax:
```python
while True:
    code1
    code2
    code3
```

# make alternate letter uppercase
1. Accept a sentence from the user.
2. Create a new variable that has an empty string
3. create a variable that tracks the index of the string which start with 0
4. Loop till we reach the end of the string
    - a. if the index is even, 
        - i. We make the character uppercase
        - ii. Add the new character to the result variable
    - b. if the index is odd,
        - i. We make the character lowercase
        - ii. Add the new character to the result variable
    - c. increase the index by 1
5. At the end of the loop print the result.

````python
sentence = input('Enter a sentence: ')
result = ''
index = 0
while index < len(sentence): 
    char = sentence[index]
    if index % 2 == 0:
        result += char.upper()
    else:
        result += char.lower()
    index += 1
print(result)
````

# for loops with lists
- We can use for loop with range and without range depending on what we want to do
- we use range function to generate indexes. This will help to edit the values in the list
- We can also directly use the list in the for loop. In this case we cannot edit the values of the list

````python
# total and average (lists/ arrays)
salesData = [5,74,53,4,78]

# Method 1: Without loop
print(sum(salesData)/ len(salesData))

# Method 2:for loop with range
total = 0
for index in range(len(salesData)): #  range(5) = [0, 1, 2, 3, 4]  index = 4
    value = salesData[index] # value = 78
    total += value  # total = 214
print(f'total :{total}')
print(f'Average : {total / len(salesData)} ')

# Method 3: for loop without range
# As we can se above, we are not editing the values in the list. So its unnecssesary to have the range function
total = 0 
for value in salesData:  #  value = 78
    total += value #  total = 214
print(f'total :{total}')
print(f'Average : {total / len(salesData)} ')  
````

````python
# adding values from two lists of different sizes.
nySales = [2,6,3,6,47,56,73,6]
dcSales = [4,6,3,6,35]
totalSales = []

# Taking care of the common part. 
length = min(len(nySales), len(dcSales))
for index in range(length):
    val1 = nySales[index]
    val2 = dcSales[index]
    totalSales.append(val1 + val2)
# taking care of remaining values
remaininValues = nySales[length: ]
totalSales.extend(remaininValues)
remaininValues = dcSales[length: ] # []
totalSales.extend(remaininValues)
print(totalSales)
````

````python
# make alternate letter uppercase
name = 'Max Winchester'
newName = ''
for index in range(len(name)):
    char = name[index]
    if index % 2 == 0:
        newName += char.upper()
    else:
        newName += char.lower()
print(newName)
````

````python
name = 'Max Winchester'
for index in range(len(name)):
    char = name[index]
    newChar = ''
    if index % 2 == 0:
        newChar = char.upper()
    else:
        newChar = char.lower()
    name = name[ :index] + newChar + name[index+1:] 
print(name)
````
