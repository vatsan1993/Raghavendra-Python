# Relationships:
- If a class is dependent on another class, there is a relationship between them.
1. has-a-relationship
2. is-a-relationship

# Has a Relationship
- if a class has another class's object as its field or member, then this relationship is called has a relationship
- Example:
    - A Car has an Engine.  (Aggregation)
        - if car is sold out, engines might not be soldout
        - weak relationship
    - A Person will have Address (Composition) 
        - if the person delete his profile, the address has to be deleted aswell.
        - strong relationship

````python
# footballplayer-height
# Note: it is better to use propeties and setters instead of getter and setter
class Height:
    def __init__(self, foot, inches):
        self._foot = foot
        self._inches = inches
    def getFoot(self):
        return self._foot
    def setFoot(self, foot):
        self._foot = foot
    def getInches(self):
        return self._inches
    def setInches(self, inches):
        self._inches = inches
    def __str__(self):
        return f"{self._foot}'{self._inches}''"
    def __repr__(self):
        return f"{self._foot}'{self._inches}''"

class FootballPlayer:
    def __init__(self, name, jersey_num, age, position, height, weight, num_games, num_wins , num_losses, num_score):
        self._name = name
        self._jersey_num = jersey_num
        self._age = age
        self._position = position
        self._height = height
        self._weight = weight
        self._num_games = num_games
        self._num_wins = num_wins
        self._num_losses = num_losses
        self._num_score = num_score
    def calc_rating(self):
        rating = 3.0
        if self._age < 35 and self._num_score > 200:
            rating = 4.2
        elif self._age < 35 and self._num_score > 80:
            rating = 3.5
        elif self._age > 35:
            if self._num_wins > 10 * self._num_losses:
                rating = 4.7
            elif self._num_wins > 5 * self._num_losses:
                rating = 4.1
            elif self._num_wins > 2 * self._num_losses:    
                rating = 3.2
        return rating
    def __str__(self):
        return f'Name= {self._name}, Age= {self._age}, Jersey= {self._jersey_num}, Height= {self._height}, Weight= {self._weight} Rating= {self.calc_rating()}'
    

h = Height(6,2)
player1 = FootballPlayer("Micheal Jordan", 10, 56, "Front-Right", h, 150, 120, 117 , 1, 520)
print(player1)
player2 = FootballPlayer("George", 21, 25, "Back-left", h, 95, 25, 15 , 3, 120)
print(player2)
````

````python
class Meeting:
    def __init__(self, meeting_id, name, description, start_time, duration, total_capacity = 100, price = 20):
        self._meeting_id = meeting_id
        self._name = name
        self._description = description
        self._start_time = start_time
        self._duration = duration
        self._end_time = start_time + timedelta(minutes=duration)
        self._total_capacity = total_capacity
        self._available_tickets = total_capacity
        self._price = price
        self._students = []
        
    def getId(self):
        return self._meeting_id
    def get_name(self):
        return self._name
    def get_description(self):
        return self._description
    def get_price(self):
        return self._price
    def __str__(self):
        return f'''Meeting_id: {self._meeting_id} 
        Name:{self._name}
        Description: {self._description}
        Start time: {self._start_time}
        End time: {self._end_time}
        Duration: {self._duration}
        Available Tickets: {self._available_tickets}
        Price: ${self._price}'''
    def __repr__(self):
        return f'''Meeting_id: {self._meeting_id} 
        Name:{self._name}
        Description: {self._description}
        Start time: {self._start_time}
        End time: {self._end_time}
        Duration: {self._duration}
        Available Tickets: {self._available_tickets}
        Price: ${self._price}'''
    def add_student(self, student):
        self._students.append(student)
        self._available_tickets -= 1
    
    def remove_student_from_meeting(self, student_id):
        for i in range(len(self._students)):
            student = self._students[i]
            if student.getId() == student_id:
                self._students.pop()
                self._available_tickets += 1
                return student
    
    def search_student_in_meeting_by_id(self, student_id):
        for student in self._students:
            if student.getId() == student_id:
                return student
        return None
    
    def search_student_in_meeting_by_phone(self, phone):
        for student in self._students:
            if student.getPhone() == phone:
                return student
        return None
    
    def search_student_in_meeting_by_Name(self, name):
        students = []
        for student in self._students:
            if name  in student.getName():
                students.append(student)
        return students
    def matching_keyword(self, keword):
        if keyword in self._name or keyword in self._description:
            return True
    def get_stundents(self):
        return self._students
````

# Is a relationship / Inheritance:
- Inheritance is a programming paradigmn where we can make two classes share some parts of code by making them parent and child.
- Child class will inherit all public properties and methods.
- private properties do not get copied down for security but we can gave getter and setter methods that can indirectly access them.
- Parent class is also called base class or super class
- Child class is also called derived class or sub class.
- Syntax:
```python
class A:
    code...

class B(A):
    code....
```

# Super keyword:
- Super referes to parent class.
- if we need to access parent class memebers then we use the super keyword.
- Some times we need to send value from child constructor to the parent constructor, so we need super method in that case
- This happens when the variables in the parent class are private.
- in the child constructor the super line has to be the first.
- Syntax:
```python
class A:
    def __init__(self, val1 , val2):
        self.__val1 = val1
        self.__val2 = val2
        
    # .. getters and setters __str__ __repr__
    
class B(A):
    def __init__(self, val1, val2, val3):
        # val1 and val2 needs to be sent to the parent
        super().__init__(val1, val2)
        self.__val3 = val3 # child's property
```

# Overriding
- Dynamic polymorphism
- In inheritance parent class method and variables will be copied down to the child class
- The parent class and the child classes will have the same exact methods.
- The parent class might have a cetain behaviour but the child class might have differnt behaviour some times.
- We need to use overriding to change the child class behaviour(method).
- Re writing a method in the child class is called overriding.



# Object Class
- A parent class to all the classes.
- object class has all the __init__, __str__, __eq__, __add__.
- We are simply overriding the behaviour.

# Types of classes
1. Concrete class - Complete class. methods inside this class will have body.
2. Abstract class - Incomplete class. methods inside the class will not have a body.

- We create an abstract class when we dont know the functionality of a method.
- Each child class will define the method in different way.
- WE cannot create an object for an abstract class.

- We need to import ABC from abc
- The class we want to make as abstract will inherit from ABC
- There can be methods which are concrete methods.
- Syntax:
```python
from abc import ABC
class MyAbstractClass(ABC):
    ....
    def method1(self):
        pass

    ....
```
